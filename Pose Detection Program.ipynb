```python
# Step 1: Import all necessary libraries
import cv2
import mediapipe as mp
```

```python
# Step 2: Identify Webcam
cap = cv2.VideoCapture(0) # 0 - Local Webcam, 1 - External Webcam
```

```python
# Leverage the Mediapipe library used for pose detection

mpPose = mp.solutions.pose
pose = mpPose.Pose()

# pose = mpPose(static_image_mode = False, upper_body_only = False, smooth_landm,arks= True, min_detection_confidence = 0.5, min_tracking_confidence = 0.5)

# To draw and connect the landmarks
mpDraw = mp.solutions.drawing_utils
```

```python
# Step 3: Switch on your webcam

while True:
    _, img = cap.read()


# Convert video/image from BGR to RGB
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


# Apply the mediapipe pose detection module for detection
    results = pose.process(imgRGB)


    print(results.pose_landmarks)


# DrawLandmarks
    if results.pose_landmarks:
        mpDraw.draw_landmarks(img, results.pose_landmarks, mpPose.POSE_CONNECTIONS)

    cv2.putText(img, "10Alytics Pose Detection" , (10, 50), cv2.FONT_HERSHEY_PLAIN, 2, (255,255,2))
    
    
    cv2.imshow("10Alytics Pose Detection", img)
    if cv2.waitKey(1) & 0xff == ord('q'):
        break

# Release the capture oce all the processing is done.
cap.release()
cv2.destroyAllWindows()
    
```
